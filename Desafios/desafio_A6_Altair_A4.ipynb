# Importando as bibliotécas



```python
import altair as alt
import pandas as pd
import numpy as np
from google.colab import drive
from IPython.display import display, HTML
```

```python
# Caminho completo do arquivo CSV no Google Drive
caminho_arquivo = '/content/drive/MyDrive/Affairs.csv'

# Lendo o arquivo CSV e carregando em um DataFrame
df = pd.read_csv(caminho_arquivo)

# Renomeando as colunas
df = df.rename(columns={
    'affairs': 'Casos', 
    'gender': 'Genero', 
    'age': 'Idade',
    'yearsmarried': 'TempoCasado',
    'children': 'Filhos',
    'religiousness': 'Religiosidade',
    'education': 'Educacao',
    'occupation': 'Ocupacao',
    'rating': 'Satisfacao'
    })
```

```python
# Divisão de fidelidade e infelidade

# Quantidade de pessoas fiéis
fidelidade = df[df["Casos"]==0]

# QUantidade de pessoas infiéis
infidelidade = df[df["Casos"]>0]
```

# Visão de fidelidade e Infidelidade por Gênero

```python
# Criando uma função para os gráficos de barras usando ALTAIR
def bar_plot(
    titulo, fidelidade, infidelidade, xlabel, xticklabels, altura, largura
    ):
    data = pd.DataFrame({
        'Categoria': xticklabels,
        'Grupos': xlabel,
        'Fidelidade': fidelidade,
        'Infidelidade': infidelidade,
    })

    fig = alt.Chart(data).mark_bar().encode(
        column="Categoria",
        x=alt.X('Barras:N', title=None),
        y=alt.Y('Valor:Q', title="Valor em [%]"),
        color=alt.Color('Barras:N', scale=alt.Scale(
            domain=['Fidelidade', 'Infidelidade'],
            range=['#1F77B4', '#D62728']
            )
        ),
    ).transform_fold(
        ['Fidelidade', 'Infidelidade'],
        as_=['Barras', 'Valor']
    )
    fig = fig.properties(
        height=altura,
        width=largura,
        title = alt.TitleParams(
            titulo,
            align = 'center'
        )
    )

    return fig
```

```python
# Gráfico de barras comparando Gênero
titulo = "Comparação de Fidelidade e Infidelidade por Gênero"
fieis = fidelidade.shape[0] / .01
infieis = infidelidade.shape[0] / .01
xticklabels = ["female","male"]
xlabel = ["Fidelidade", "Infidelidade"]
altura = 300
largura = 275

hist_ff = round(
    (fidelidade["Genero"].value_counts()[xticklabels] / fieis)
    * 100, 2
    )

hist_uff = round(
    (infidelidade["Genero"].value_counts()[xticklabels] / infieis)
    * 100, 2
    )

xticklabels = [_.title() for _ in xticklabels]

fig = bar_plot(titulo, hist_ff, hist_uff, xlabel, xticklabels, altura, largura)

fig
```

* Gênero - Assumindo que o conjunto de dados representa uma amostra aleatória da população, se não houver correlação entre sexo e fidelidade, a proporção de mulheres para homens deve ser semelhante entre os grupos "fiéis" e "infiéis".

* Da mesma forma, se houver correlação, o equilíbrio de gênero deve diferir entre os dois grupos. Indo do grupo "fiel" para o grupo "infiel", o equilíbrio de gênero se inclina para o sexo masculino, sugerindo que os homens podem estar mais inclinados a trair.



# Visualização por faixas de idade

```python
# Gráfico de barras comparando a fidelidade por faixas de idade
titulo = "Comparação de Fidelidade e Infidelidade por Faixa Etária"
xlabel = [0,17.5,22,27,32,37,42,47,52,57]
xticklabels = [
    "- 20","20-24","25-29",
    "30-34","35-39","40-44",
    "45-49","50-54","> 54"
    ]
altura = 300
largura = 50

qtd_fieis = pd.cut(fidelidade['Idade'], bins=xlabel, labels=xticklabels)
hist_ff = qtd_fieis.value_counts(sort=False) / fieis

qtd_infieis = pd.cut(infidelidade['Idade'], bins=xlabel, labels=xticklabels)
hist_uff = qtd_infieis.value_counts(sort=False) / infieis

bar_plot(titulo, hist_ff, hist_uff, 'Idade', xticklabels, altura, largura)
```

* Idade - Seguindo o mesmo argumento, se não houver correlação entre idade e fidelidade, a distribuição etária deveria ser semelhante entre os grupos "fiéis" e "infiéis" e diferente nos demais. 

* Indo do grupo "fiel" para o grupo "infiel", a distribuição se inclina para idades mais altas, sugerindo que indivíduos mais velhos podem estar mais inclinados a trapacear.



# Visão comparativa de fidelidade por:

* Tempo de casamento
* Se o casal possui filhos
* Nível de religiosidade do casal
* Satisfação no relacionamento


```python
# Gráfico 1 - de barras por anos de casamento
titulo1 = "Fidelidade e Infidelidade por Anos de Relacionamento"
faixas_casamento = [0,.75,1.5,4,7,10,15]
xticklabels_tcasamento = [
    "- 1", "1-2",
    "3-5", "6-8",
    "9-11", "> 11"
    ]
altura1 = 250
largura1 = 60

qtd_fieis_tcasamento = pd.cut(fidelidade["TempoCasado"],
                              bins = faixas_casamento, 
                              labels = xticklabels_tcasamento)

hist_ff_tcasamento = qtd_fieis_tcasamento.value_counts(
    sort=False
    ) / fieis

qtd_infieis_tcasamento = pd.cut(infidelidade["TempoCasado"],
                                bins = faixas_casamento,
                                labels = xticklabels_tcasamento)

hist_uff_tcasamento = qtd_infieis_tcasamento.value_counts(
    sort=False
    ) / infieis


# Gráfico 2 - de barras por casais com ou sem filhos
titulo2 = "Fidelidade e Infidelidade por Casais com ou sem Filhos"
xticklabels_filhos = ["no","yes"]
altura2 = 250
largura2 = 227

hist_ff_filhos = fidelidade["Filhos"].value_counts(
    
)[xticklabels_filhos] / fieis
hist_uff_filhos = infidelidade["Filhos"].value_counts(
    
)[xticklabels_filhos] / infieis

xticklabels_filhos = [_.title() for _ in xticklabels_filhos]


# Gráfico 3 - de visualização por da religiosidade dos casais
titulo3 = "Fidelidade e Infidelidade por Religiosidade"
nivel_religiosidade = list(range(6))
xticklabels_religiosidade = [
    "1-Anti","2-Not At All","3-Slightly","4-Somewhat","5-Very"
    ]
altura3 = 250
largura3 = 76

qtd_fieis_religiosidade = pd.cut(fidelidade["Religiosidade"], 
                   bins = nivel_religiosidade, 
                   labels=xticklabels_religiosidade)

hist_ff_religiosidade = qtd_fieis_religiosidade.value_counts(
    sort=False
    ) / fieis

qtd_infieis_religiosidade = pd.cut(infidelidade["Religiosidade"], 
                     bins = nivel_religiosidade, 
                     labels=xticklabels_religiosidade)

hist_uff_religiosidade = qtd_infieis_religiosidade.value_counts(
    sort=False
    ) / infieis


# Gráfico 4 - de comparação por nível de felicidade no matrimônio
titulo4 = "Fidelidade e Infidelidade por Felicidade no Matrimônio"
nivel_satisfacao = list(range(6))
xticklabels_satisfacao = [
    "1-Very\nUnhappy",
    "2-Somewhat\nUnhappy",
    "3-Average",
    "4-Happier\nThan\nAverage",
    "5-Very\nHappy"
    ]
altura4 = 250
largura4 = 76

qtd_fieis_satisfacao = pd.cut(fidelidade["Satisfacao"], 
                   bins = nivel_satisfacao, 
                   labels = xticklabels_satisfacao)

hist_ff_satisfacao = qtd_fieis_satisfacao.value_counts(
    sort=False
    ) / fieis

qtd_infieis_satisfacao = pd.cut(infidelidade["Satisfacao"], 
                     bins = nivel_satisfacao, 
                     labels = xticklabels_satisfacao)

hist_uff_satisfacao = qtd_infieis_satisfacao.value_counts(
    sort=False
    ) / infieis
```

```python
# Plotando os 4 gráficos
# Gráfico 1
chart1 = bar_plot(titulo1, hist_ff_tcasamento, hist_uff_tcasamento, "Gráfico 1", xticklabels_tcasamento, altura1, largura1)

# Gráfico 2
chart2 = bar_plot(titulo2, hist_ff_filhos, hist_uff_filhos, "Gráfico 2", xticklabels_filhos, altura2, largura2)

# Gráfico 3
chart3 = bar_plot(titulo3, hist_ff_religiosidade, hist_uff_religiosidade, "Gráfico 3", xticklabels_religiosidade, altura3, largura3)

# Gráfico 4
chart4 = bar_plot(titulo4, hist_ff_satisfacao, hist_uff_satisfacao, "Gráfico 4", xticklabels_satisfacao, altura4, largura4)

# Concatenação vertical dos dois gráficos superiores e dos dois inferiores
top_row = alt.hconcat(chart1, chart2)
bottom_row = alt.hconcat(chart3, chart4)

# Concatenação horizontal dos dois grupos de gráficos
combined_chart = alt.vconcat(top_row, bottom_row)

#Exibir o gráfico combinado
combined_chart
```

* Anos de casamento, filhos, religiosidade e felicidade conjugal - Estendendo esta análise aos demais atributos, os "infiéis" são casados há mais tempo, têm mais filhos, são menos religiosos e menos felizes com o casamento do que os "fiéis"

* Com isso é sugerível que indivíduos com essas características podem estar mais inclinadas a trapacear.

```python
# Gráfico de visualização por nível de ocupação (classificação de Hollingshead)
titulo = "Fidelidade e Infidelidade por Ocupação Social"
xticklabels=list(range(1,8))
altura = 300
largura = 100

hist_ff = fidelidade["Ocupacao"].value_counts()[xticklabels] / fieis
hist_uff = infidelidade["Ocupacao"].value_counts()[xticklabels] / infieis

xlabel=\
"Ocupação (Classificação por Posição Social em Ordem Asc)"

bar_plot(titulo, hist_ff, hist_uff, xlabel, xticklabels, altura, largura)
```

* Educação e ocupação - Embora a distribuição de educação e ocupação seja diferente entre os grupos "fiéis" e "infiéis", ao contrário do que observamos para os outros atributos, não há padrões óbvios. 


* Isso implica que a fidelidade pode estar correlacionada com a educação e a ocupação, mas é obscurecida pela forma como a educação e a ocupação são categorizadas e classificadas no conjunto de dados. 


* No caso da ocupação, o conjunto de dados faz uso da pontuação de ocupação de Hollingshead, que classifica as ocupações por posição social, em vez de outros fatores, como salário ou jornada de trabalho, que podem ser mais relevantes para a fidelidade.

```python
import altair as alt
from vega_datasets import data

database = df

slider = alt.binding_range(min=0, max=60, step=5)
select_year = alt.selection_single(name='Idade', fields=['Idade'],
                                   bind=slider, init={'Idade': 30})

base = alt.Chart(database).add_selection(
    select_year
).transform_filter(
    select_year
).transform_calculate(
    gender=alt.expr.if_(alt.datum.Genero == 'male', 'Homens', 'Mulheres')
).properties(
    width=250
)

color_scale = alt.Scale(domain=['Homens', 'Mulheres'],
                        range=['#1f77b4', '#e377c2'])

left = base.transform_filter(
    alt.datum.gender == 'Mulheres'
).encode(
    y=alt.Y('Idade:O', axis=None),
    x=alt.X('count():Q',
            title='Mulheres infiéis',
            sort=alt.SortOrder('descending')),
    color=alt.Color('gender:N', scale=color_scale, legend=None)
).mark_bar().properties(title='Mulheres')

middle = base.encode(
    y=alt.Y('Idade:O', axis=None),
    text=alt.Text('Idade:Q'),
).mark_text().properties(width=30)

right = base.transform_filter(
    alt.datum.gender == 'Homens'
).encode(
    y=alt.Y('Idade:O', axis=None),
    x=alt.X('count():Q', title='Homens infiéis'),
    color=alt.Color('gender:N', scale=color_scale, legend=None)
).mark_bar().properties(title='Homens')

alt.concat(left, middle, right, spacing=5)
```
