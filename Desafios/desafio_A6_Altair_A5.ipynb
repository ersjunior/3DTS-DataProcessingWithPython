## Iniciando a análise dos dados

```python
# Caminho completo do arquivo CSV no Google Drive
caminho_arquivo = '/content/drive/MyDrive/ds_salaries.csv'

# Lendo o arquivo CSV e carregando em um DataFrame
df = pd.read_csv(caminho_arquivo)

# Renomeando as colunas
df = df.rename(columns={
    'work_year': 'Ano_Trabalho',
    'experience_level': 'Nivel_Experiencia',
    'employment_type': 'Tipo_Registro',
    'job_title': 'Cargo',
    'salary': 'Salario',
    'salary_currency': 'Moeda',
    'salary_in_usd': 'Salario_Dolar',
    'employee_residence': 'Local_Funcionario',
    'remote_ratio': 'Contrato_Remoto',
    'company_location': 'Local_Empresa',
    'company_size': 'Tamanho_Empresa',
})

# Visualisando o dataframe + dados
print(df.shape)
df.head(5)
```

## Verificando o volume de profissionais e seus níveis de experiência

```python
# Renomeando os campos da coluna de nível
df['Nivel_Experiencia'] = df['Nivel_Experiencia'].replace({
    'EN': 'Junior',
    'MI': 'Pleno',
    'SE': 'Senior',
    'EX': 'Executive'
})

# Criação do gráfico com Altair
ex_level = df['Nivel_Experiencia'].value_counts().reset_index().rename(
    columns={
        'index': 'Nivel',
        'Nivel_Experiencia': 'Qtd'
    }
)
# Combinação dos campos "nivel" e "contagem" em uma única coluna "rotulo"
ex_level['rotulo']=ex_level['Nivel']+" ("+ex_level['Qtd'].astype(str)+")"
print(ex_level[['Nivel', 'Qtd']])


base = alt.Chart(ex_level).encode(
    theta=alt.Theta("Qtd:Q", stack=True),
    radius=alt.Radius(
        "Qtd", 
        scale=alt.Scale(
            type="sqrt", 
            zero=True, 
            rangeMin=20
            )
        ),
    color="Nivel:N",
).properties(
    width=800,
    height=400,
)

c1 = base.mark_arc(innerRadius=20, stroke="#fff")

c2 = base.mark_text(radiusOffset=30, fontWeight='bold').encode(
    text="rotulo:N",
)

c1 + c2
```

* De acordo com a amostra, a maioria dos trabalhadores estão em nível de Sênior / Especialista.
Sendo um volume considerável em relação aos iniciantes / Junior's e com um número muito pequeno quando se trata de cargos em nível executivo

* Um detalhe curioso é que o volume de profissionais sêniores nos cargos, são maiores que a quantidade de todos os outros cargos somados

## Vendo os cargos de trabalho mais frequentes

```python
# Criando uma variável que vai receber os 10 cargos com maior frequência
top10_cargos = df['Cargo'].value_counts()[:10].reset_index().rename(
    columns={
        'index': 'Cargo',
        'Cargo': 'Qtd'
    }
)
print(top10_cargos.head(3))
cor_hex = '#636EFA'

# Criação do gráfico com Altair
fig = alt.Chart(top10_cargos).mark_bar().encode(
    x=alt.X(
        'Cargo:N', 
        title=None, 
        sort=alt.EncodingSortField(
            field="Qtd", 
            op="sum", 
            order="descending"
            )
        ),
    y=alt.Y('Qtd:Q', title="Quantidade"),
    color=alt.value(cor_hex),
).properties(
    title="Top 10 Cargos",
    width=1000,
    height=400,
)

text = fig.mark_text(
    align='center',
    baseline='bottom',
    fontSize=15,
    fontWeight='bold',
    dy=-5    
).encode(
    text='Qtd'
)

# Mostrando o gráfico
fig + text
```

* Os 3 cargos mais comuns na área de dados são

* 1) Engengeiro de dados
* 2) Ciêntista de dados
* 3) Analista de dados

## Temos 4 tipos de Empregos

* PT: Meio período (Part-time)

* FT: Tempo integral (Full-time)

* CT: Contrato (Contract)

* FL: Autônomo (Freelance)

```python
# Trazendo o volume total de tipo de trabalho e grupos 
df['Tipo_Registro'] = df['Tipo_Registro'].replace({
    'FT': 'Integral',
    'PT': 'Meio Período',
    'CT': 'Contrato',
    'FL': 'Freelance'
})
grupo = df['Tipo_Registro'].value_counts().reset_index().rename(
    columns={
        'index': 'Registro',
        'Tipo_Registro': 'Qtd'
    }
)
cor_hex = '#EF553B'
print(grupo)

# Criação do gráfico com Altair
fig = alt.Chart(grupo).mark_bar().encode(
    x = alt.X(
        'Registro:N',
        title=None,
        sort=alt.EncodingSortField(
            field="Qtd",
            op='sum',
            order='descending',
        )
    ),
    y = alt.Y(
        'Qtd:Q',
        title='Quantidade'
    ),
    color=alt.value(cor_hex),
).properties(
    title="Distribuição - Tipo de trabalho",
    width=1000,
    height=400, 
)

# Definições do gráfico
text = fig.mark_text(
    align='center',
    baseline='bottom',
    fontSize=15,
    fontWeight='bold',
    dy=-5,
).encode(
    text='Qtd:Q'
)

# Mostrando o gráfico
fig + text
```

Conforme a distribuição, percebemos que:

* A maior recorrência de contratos é em período integral

* Pouquissimos casos são freelance e/ou meio período

## Comparação de local de empresas e residência de funcionários 

```python
# Definindo a localidade de empresas / residência de funcionários
paisEmp = coco.convert(
    names=df['Local_Empresa'], 
    to="ISO3"
)
df['Local_Empresa'] = paisEmp

paisFunc = coco.convert(
    names=df['Local_Funcionario'],
    to='ISO3'
)
df['Local_Funcionario'] = paisFunc

local_empresa = df['Local_Empresa'].value_counts(
).reset_index(
).rename(
    columns={
        'index': 'Local',
        'Local_Empresa': 'Qtd'
    }
)
top10_local_empresa = local_empresa[:10]
top10_local_empresa['Tipo'] = 'Local Empresa'

local_func = df['Local_Funcionario'].value_counts(
).reset_index(
).rename(
    columns={
        'index': 'Local',
        'Local_Funcionario': 'Qtd'
    }
)
top10_local_func = local_func[:10]
top10_local_func['Tipo'] = 'Local Funcionário'

base = pd.concat([top10_local_empresa, top10_local_func])
print(base.head(3))


# Criação do gráfico com Plotly Express
fig = alt.Chart(base).mark_text(
    align='center',
    baseline='bottom',
    fontSize=11,
    fontWeight='bold',
    radiusOffset=30,
    dy=-10,
).mark_bar(
).encode(
    x = alt.X(
        'Tipo:O',
        title=None,
        sort=alt.EncodingSortField(
            field='Qtd',
            op='sum',
            order='descending',
        ),
    ),
    y = alt.Y(
        'sum(Qtd):Q',
        title='Quantidade',
    ),
    color='Tipo:N',
    column= alt.Column(
        'Local:N',
        title=None,
        sort=alt.EncodingSortField(
            field='Qtd',
            op='sum',
            order='descending',
        ),
    ),
    text=alt.Text('Qtd:N', format='.0f'),
).properties(
    title="Comparação da Residência do Funcionário e Local da Empresa",
    width=50,
    height=400,
    config=custom_config
)
# Mostrando o gráfico
fig
```

* Em 2023, todos os países possuem um número semelhante de residências de funcionários e localizações de empresas.

## Distribuição por tamanho das empresas

* S = Empresas de pequeno porte

* M = Empresas de médio porte

* L = Empresas de grande porte

```python
# Definindo os grupos de empresas
grupos = df['Tamanho_Empresa'].value_counts().reset_index().rename(
    columns={
        'index': 'Tamanho',
        'Tamanho_Empresa': 'Qtd',
    }
)
grupos['Tamanho'] = grupos['Tamanho'].replace({
    'S': 'Pequeno Porte',
    'M': 'Médio Porte',
    'L': 'Grande Porte',
})
print(grupos.head(3))

# Criação do gráfico com Plotly Express
fig = alt.Chart(grupos).mark_bar().encode(
    x = alt.X(
        'Tamanho:N',
        title=None,
        sort=alt.EncodingSortField(
            field="Qtd",
            op='sum',
            order='descending',
        )
    ),
    y = alt.Y(
        'Qtd:Q',
        title='Quantidade'
    ),
    color = 'Tamanho:N',
).properties(
# Definições do gráfico
    title="Distribuição de Funcionários por porte da Empresa",
    # Largura (w) e altura (h) em pixels
    width=1000,
    height=400,
)

# Definições do gráfico
text = fig.mark_text(
    align='center',
    baseline='bottom',
    fontSize=15,
    fontWeight='bold',
    dy=-5,
).encode(
    text="Qtd:Q"
)
# Mostrando o gráfico
fig + text
```

* O tamanho das empresas consistem principalmente em empresas de médio porte, em seguida, as de grande porte, ficando por ultimo as de pequeno porte.

## Análise da distribuição de funcionários trabalhando Remoto, Hibrido e Presencial

A coluna 'remote_ratio' consiste em 3 valores:

* 0: Nenhum trabalho remoto (menos de 20%)

* 50 : Parcialmente remoto

* 100 : Totalmente remoto (mais de 80%)

```python
# Trazendo o volume total de tipo de trabalho e grupos 
df['Contrato_Remoto'] = df['Contrato_Remoto'].replace({
    100: 'Remoto',
    50: 'Hibrido',
    0: 'Presencial',
})

# Criando o gráfico
fig = alt.Chart(df).mark_bar().encode(
    x = alt.X(
        'Contrato_Remoto',
        title=None,
        sort=alt.EncodingSortField(
            field="Salario_Dolar",
            op='sum',
            order='descending',
        ),
    ),
    y = alt.Y(
        'count(Contrato_Remoto):Q',
        title='Quantidade',
    ),
    color = 'Contrato_Remoto:N',
).properties(
    title = 'Distribuição do tipo de trabalho',
    # Largura (w) e altura (h) em pixels
    width=1000,
    height=400,
)
# Definições do gráfico
text = fig.mark_text(
    align='center',
    baseline='bottom',
    fontSize=15,
    fontWeight='bold',
    dy=-5,
).encode(
    text='count(Contrato_Remoto):Q'
)

# Mostrando o gráfico
fig + text
```

* Em 2023 a maioria das empresas ainda seguirá a rota totalmente remota, seguida de perto pela política híbrida/parcialmente remota e, em seguida, pela política sem controle presencial.
